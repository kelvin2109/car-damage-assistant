## **Conclusion**
The dataset contains five main keys: **licenses**, **info**, **categories**, **images**, and **annotations**.  
For car damage detection, the **important components** are `images`, `annotations`, and `categories`.  
- `annotations` provide bounding boxes (`bbox`) and category IDs — essential for training.  
- `segmentation` and `attributes` inside `annotations` are **not required** unless we do segmentation tasks.  
- `licenses` and `info` are **metadata only**, not used for model training.

Additional dataset characteristics:
- Image resolutions are generally consistent (e.g., around **1000×750**), providing enough detail for detection.  
- With **2,816 images** and **6,211 annotations**, each image contains on average **~2 damage instances**.

```python
def clean_annotations(data):
    
    #path for input annotation-file
    input_path = "data/annotations/instances_train.json"
    #path for cleaned annotation-folder
    cleaned_dir = "data/annotations/cleaned/"
    clean_file_name = "annotations_train_cleaned.json"
    #Create folder if it doesn’t exist
    os.makedirs(cleaned_dir,exist_ok=True)
    #check if we already had the cleaned annotation file'
    print(os.path.join(cleaned_dir,clean_file_name))
    if os.path.exists(os.path.join(cleaned_dir,clean_file_name)):
        print("The cleaned file already exist.")
        
        with open(os.path.join(cleaned_dir,clean_file_name),"r") as f:
            data= json.load(f)
    
    else:
        with open(input_path,"r") as f:
            data = json.load(f)
        
        #Remove unnecessary meta-data
        keys = ["licenses","info"]
        for key in keys:
            if key in data:
                del data[key]
                print(f"{key} successfully deleted!")
                
        #Remove unnecessary annotation-field
        fields_to_remove = ["segmentation","attributes"]
        for del_ in fields_to_remove:
            deleted = False
            for anno in data["annotations"]:
                if del_ in anno:
                    anno.pop(del_,None)
                    deleted = True
            if(deleted):
                print(f"{del_} successfully deleted!")
                
        with open(os.path.join(cleaned_dir,clean_file_name),"w") as f:
            json.dump(data,f,indent = 2)
            
        print(f"Cleaned file saved at {os.path.join(cleaned_dir,clean_file_name)}")
                    
    return data
```

```python
data = clean_annotations(data)
```

## Class Distribution Conclusion

The dataset displays a **clear class imbalance** across the six damage categories.  
The classes **scratch (2560 samples)** and **dent (1806 samples)** are significantly more frequent than the others.  
In contrast, **tire flat (225 samples)**, **glass shatter (475 samples)**, and **lamp broken (494 samples)** have much fewer samples.

This imbalance may cause the model to **bias predictions toward the common classes**, leading to **lower recall and poor mAP** for the minority classes.

### Recommended Measures
We will address this imbalance by applying **class-weighted loss** in the **classification head of Faster R-CNN**.  
This approach increases the contribution of minority classes during training, helping the model learn to recognize them more effectively without altering the dataset itself.

### Evaluation Focus
- Monitor **per-class mAP**, not only overall mAP, to ensure performance is balanced across all damage categories.

## Prepare dataframe for multi-class image sample 

```python
#Keep all duplicated annotations (images with more than one object)
duplicated_ann = df_ann[df_ann.duplicated(subset=["image_id"], keep=False)]
```

```python
duplicated_ann
```

```python
def get_image_per_category(col1,col2):
    
    seen1 = set()
    seen2 = set()
    mask = []

    for c1, c2 in zip(col1, col2):
        both_new = (c1 not in seen1) and (c2 not in seen2)
        mask.append(both_new)   #~flips False → True, True → False

        if(both_new):
            seen1.add(c1)
            seen2.add(c2)
    return mask
```

```python
multiclass_img = duplicated_ann[get_image_per_category(duplicated_ann["category_id"],duplicated_ann["image_id"])].drop(columns=["id"]).reset_index(drop=True)
```

```python
multiclass_img
```

```python
all_per_category_dup = (df_ann[df_ann["image_id"].isin(multiclass_img["image_id"])].drop(columns=["id"])
                        .merge(df_cat,left_on = "category_id",right_on = "id").drop(columns=["id"]))
```

```python
# Aggregate all categories, areas, and bounding boxes for each duplicated image
agg_ann_per_image = all_per_category_dup.groupby("image_id").agg({'category_id':list , 'bbox':list, "name":lambda x: list(set(x)) }).reset_index()
```

```python
def show_presentated_image_eachclass( sample_img, num_class, img_dir="data/train/", fig_size =(10,10)):

    """
    Display sample images per class with bounding boxes for EDA.

    Parameters
    ----------
    sample_img : pd.DataFrame
        DataFrame with columns ["file_name", "bbox", "name"].
    num_class : int
        Number of classes to display.
    img_dir : str
        Directory containing image files.
    color : tuple
        BGR color for bounding boxes.
    figsize : tuple
        Figure size for the plot.
    """
    
    # Convert to list of dicts for easy iteration
    sample_img = sample_img[["file_name","category_id","bbox","name"]].to_dict(orient="records")

    # Create subplot grid for showing images
    row = math.ceil(num_class/3)
    fig, axes = plt.subplots(row,3,figsize=(10,10))
    fig.suptitle("Sample Images per Class", fontsize=16., y=0.9)
    plt.subplots_adjust(wspace=0.1, hspace=0.1)
    
    # List to store images after loading and drawing bbox
    images = []
    
    # Select 1 sample per class
    for path in sample_img:
        img = os.path.join("data/train/", path['file_name'])
        img = cv2.imread(img)
        color = (0, 255, 0)#RGB
        thickness = 2
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        if isinstance(path["category_id"], list):
            for i in path["bbox"]:
                x, y, w, h = i
                cv2.rectangle(img, (int(x), int(y)), (int(x + w), int(y + h)), color, thickness)
            images.append(img)
        else:
            x, y, w, h = path["bbox"]
            cv2.rectangle(img, (int(x), int(y)), (int(x + w), int(y + h)), color, thickness)
            images.append(img)
    
    # Display each image with its class title
    for i, ax in enumerate(axes.flat):
        ax.set_title("Class: {}".format(sample_img[i]["name"]), fontsize=12)
        if i < len(images):
            ax.imshow(images[i])
        ax.axis('off')
```
